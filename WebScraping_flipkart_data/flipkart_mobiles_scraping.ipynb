```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
import re
```

```python
ProductName=[]
Prices=[]
Description=[]
Reviews=[]
for i in range(2,12):
    if i == 4:
        continue


    url="https://www.flipkart.com/search?q=mobiles+under+50000+rupees&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    print(url)
    r=requests.get(url).text
    #print(r)
    soup=BeautifulSoup(r,"lxml")
    data=[]
    box=soup.find("div",class_="_1YokD2 _3Mn1Gg")



    names=box.find_all("div",class_="_4rR01T")
    #print(names)
    for i in names:
        name=i.text
        ProductName.append(name)

    print(len(ProductName))



    prices=box.find_all("div",class_="_30jeq3 _1_WHN1")
    #print(names)
    for i in prices:
        price=i.text
        price = re.sub('[^0-9]', '', price)
        Prices.append(price)

    print(len(Prices))

    descriptions=box.find_all("ul",class_="_1xgFaf")
    #print(names)
    for i in descriptions:
        description=i.text
        Description.append(description)

    print(len(Description))


    reviews=box.find_all("div",class_="_3LWZlK")

    #print(names)
    for i in reviews:
        review=i.text
        if len(review) > 0:
            Reviews.append(review)
        else:
            Reviews.append(0)


    #     # Process the elements
    #     for i in reviews:
    #         review=i.text
    #         print(review)
    #         Reviews.append(review)
    #         # Do something with the value
    # else:
    #     # Handle the case when no elements are found
    #     Reviews.append(0)

    #     Reviews.append(review)

    #     # Check if any elements are found
    # if len(reviews) > 0:
    #     # Process the elements
    #     for i in reviews:
    #         review=i.text
    #         print(review)
    #         Reviews.append(review)
    #         # Do something with the value
    # else:
    #     # Handle the case when no elements are found
    #     Reviews.append(0)


    # #print(names)
    # for i in reviews:
    #     review=i.text
    #     Reviews.append(review)
    

    print(len(Reviews))
    # nextPage=soup.find("a",class_="_1LKTO3")["href"]
    # # print(nextPage)
    # completeNextPage="https://www.flipkart.com"+nextPage
    # print(completeNextPage)



df=pd.DataFrame({"Product Name":ProductName,"Prices":Prices,"Description":Description,"Reviews":Reviews})
print(df)

df.to_csv("F:/SEM9/PCD LAB/Web_scraping/flipkart/flipkart_mobiles_under_50000.csv")
# Specify the desired filename for the CSV file

# filename = 'flipkart.csv'
# data.append({"Product Name":ProductName,"Prices":Prices,"Description":Description,"Reviews":Reviews})

# # Save the data as a CSV file
# with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
#     fieldnames = ['Product Name', 'Prices','Description','Reviews']
#     writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

#     writer.writeheader()
#     for item in data:
#         writer.writerow(item)



```
