```python
import pandas as pd
```

```python
us_babies=pd.read_csv("us_baby_names.csv")
us_babies
```

```python
us_babies['Year']
```

```python
#selecting only 2014 year data
us_babies_2014=us_babies.loc[us_babies['Year']==2014,:]

```

```python
us_babies_2014
```

```python
# sorting the data based on count -descending
us_babies_2014_sorted=us_babies_2014.sort_values('Count',ascending=False)
us_babies_2014_sorted
```

```python
#Selecting top 5 using iloc
us_babies_2014_sorted.iloc[0:5]
```

Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listings=pd.read_csv('listings.csv')
listings
```

```python
sn.countplot(x='neighbourhood_group',data=listings)
plt.show()
```

```python
sn.barplot(x='neighbourhood_group',y='price',data=listings)
plt.show()
```

```python
sn.barplot(x='neighbourhood_group',y='price',ci=False,data=listings)
plt.show()
```

```python
import numpy as np
plt.hist(listings['price'],bins=np.arange(0,1100,40))
plt.xlabel('Price in US dollars')
```

```python
plt.scatter(x=listings['price'],y=listings['number_of_reviews'])
plt.xlabel('Price')
plt.ylabel('number_of_reviews')
plt.show()
```

```python
plt.scatter(x=listings['price'],y=listings['number_of_reviews'])
plt.xlabel('Price')
plt.ylabel('number_of_reviews')
plt.xlim(0,1000)
plt.show()
```
