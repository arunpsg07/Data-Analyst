```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv(r'F:\important\Data analyst\Projects\Customer segmentation -Kmeans\Dataset\Mall_Customers.csv')
df.head(3)
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,[3,4]].values
```

```python
print(x)
```

```python
# Finding optimal number of clusters
# WCSS-->within clusters sum of squares 
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title("The elbow pint graph")
plt.xlabel("Number of clusters")
plt.ylabel("wcss values")
plt.show()
```

At 5 there is no significant drop after that 
So optimun number of clusters is 5

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=42)
```

```python
#label for each clusters
y=kmeans.fit_predict(x)
print(y)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
